# Derive Notebook

A symbolic mathematics environment for Python.

```
x = Symbol('x')
```

```
# Symbolic differentiation
D(Sin(x) * Exp(x), x)
```

```
# Integration
Integrate(x**2 * Exp(-x), x)
```

```
# Plotting
Plot([Sin(x), Cos(x)], (x, 0, 2*Pi))
```
